```python
import pulp
lp = pulp.LpProblem("LPP1", pulp.LpMaximize)
x = pulp.LpVariable('x', lowBound=0)
y = pulp.LpVariable('y', lowBound=0)
lp += 3*x + 2*y
lp += x + y <= 4
lp += x <= 2
lp += y <= 3
lp.solve()
print(x.value(), y.value(), pulp.value(lp.objective))
```
